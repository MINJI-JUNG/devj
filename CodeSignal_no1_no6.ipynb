## 2. Given a year, return the century it is in. The first century spans from the year 1 up to and including the year 100, the second - from the year 101 up to and including the year 200, etc.

### Example
* For year = 1905, the output should be centuryFromYear(year) = 20;
* For year = 1700, the output should be centuryFromYear(year) = 17.

### Input/Output
* [input] integer year :: A positive integer, designating the year. 1≤year≤2005.
* [output] integer :: The number of the century the year is in.

### 2.1 My Answer

```python
import numpy

def centuryFromYear(year):
    max_century = 30
    all_century = list(range(1, max_century+1))
    all_year    = list(numpy.arange(0, (max_century+1)*100, 100))

    loop = [x-1 for x in all_century]
    
    for i in loop:
        if all_year[i] < year <= all_year[i+1]:
            find_century = all_century[i]
    return find_century
```

```python
centuryFromYear(1905)
```

## 3. Given the string, check if it is a palindrome.

### Example
* For inputString = "aabaa", the output should be checkPalindrome(inputString) = true;
* For inputString = "abac", the output should be checkPalindrome(inputString) = false;
* For inputString = "a", the output should be checkPalindrome(inputString) = true.

### Input/Output
* [input] A non-empty string consisting of lowercase characters.
* [output] boolean

### 3.1 My Answer

```python
def checkPalindrome(inputString):
  num_char = len(inputString)
  
  if num_char%2 == 0:  # Even number
    newchar = list(inputString)
  else:  # Odd number: delete middle character
    mid_char    = num_char//2
    newchar     = list(inputString[:mid_char] + inputString[mid_char+1:])
  
  newchar_len = len(newchar)
  check = []

  for i in list(range(0,newchar_len)):
    left_char  = newchar[i]
    right_char = newchar[newchar_len-i-1]
    # check a palindrome
    if left_char != right_char:
      check.append(False)
      break
    else:
      check.append(True)
      if i == (newchar_len/2)-1:
        break

  return(all(check))
```

```python
checkPalindrome("hlbeeykoqqqqokyeeblh")
```

## 4. Given an array of integers, find the pair of adjacent elements that has the largest product and return that product. 
* 인접한 두 정수의 곱 중에서 가장 큰 값을 리턴하라

### Example
* For inputArray = [3, 6, -2, -5, 7, 3], the output should be adjacentElementsProduct(inputArray) = 21.
* 7 and 3 produce the largest product.

### Input/Output
* [input] 2 ≤ inputArray.length ≤ 10,  -1000 ≤ inputArray[i] ≤ 1000
* [output] The largest product of adjacent elements.

### 4.1 My Answer

```python
def adjacentElementsProduct(inputArray):
    products = []
    for i in range(1,len(inputArray)):
        products.append(inputArray[i-1]*inputArray[i])
    return(max(products))
```

```python
inputArray = [3, 6, -2, -5, 7, 3]
adjacentElementsProduct(inputArray)
```

## 5. Below we will define an n-interesting polygon. Your task is to find the area of a polygon for a given n. An n-interesting polygon is obtained by taking the n - 1-interesting polygon and appending 1-interesting polygons to its rim, side by side. You can see the 1-, 2-, 3- and 4-interesting polygons in the picture below.

```python
from IPython.display import Image
Image(filename='D:/[Python]CodeSignal/no_5_image.png', width=400)
```

### Example
* For n = 2, the output should be shapeArea(n) = 5
* For n = 3, the output should be shapeArea(n) = 13

### Input/Output
* [input] integer n
* [output] integer (The area of the n-interesting polygon)

### 5.1 My Answer

```python
def shapeArea(n):
    cum_n = list(range(1, n+1))  # for cumulating
    new_area = [i+(i-1)+(i-1)+(i-2) for i in cum_n]  # new area to its rim, 4-side
    final_area = 1 + sum(new_area)  # 1 = the area of 1-interesting polygon
    return(final_area)

shapeArea(5)
```

### 5.2 Best Score
* 나와 최고점자의 차이점
    * 나: 새롭게 더해지는 가장자리 네 군데 면적을 계산한 후, 누적합으로 최종 면적 산출
    * 최고점자: 1, 5, 13, 25, ... 로 늘어나는 면적 계산식을 만들어버림. (만약 계산식의 수식을 시간들여 꼼꼼히 봤으면 찾았을 수도..)
   
* 최고점자 따라잡기
    * n=1 : 1
    * n=2 : 1+4      → 1+4  → $1+(2*2)*(2-1)$
    * n=3 : 1+4+8    → 1+12 → $1+(2*3)*(3-1)$
    * n=4 : 1+4+8+12 → 1+24 → $1+(2*4)*(4-1)$
    * n   : $1+(2*n)*(n-1)$

```python
def shapeArea(n):
    return 1+(2*n)*(n-1)

shapeArea(5)
```

## 6. Ratiorg got statues of different sizes as a present from CodeMaster for his birthday, each statue having an non-negative integer size. Since he likes to make things perfect, he wants to arrange them from smallest to largest so that each statue will be bigger than the previous one exactly by 1. He may need some additional statues to be able to accomplish that. Help him figure out the minimum number of additional statues needed.
* 리스트 안의 원소 배열이 1씩 커지도록 하기 위해, 필요한 최소한의 원소 개수를 리턴하라

### Example
* For statues = [6, 2, 3, 8], the output should be makeArrayConsecutive2(statues) = 3.
* 필요한 원소 개수는 3개 (= 4, 5, 7)

### Input/Output
* [input] 1 ≤ statues.length ≤ 10, 0 ≤ statues[i] ≤ 20
* [output] The minimal number of statues

### 6.1 My Answer
* 필요한 원소 개수(need_element)를 찾고, 현재 원소 개수(now_element)를 빼준다

```python
def makeArrayConsecutive2(statues):
    need_element = max(statues)-min(statues)+1
    now_element = len(statues)
    return(need_element-now_element)
```

```python
statues = [6, 2, 3, 8]
makeArrayConsecutive2(statues)
```

### 6.2 Best Score
